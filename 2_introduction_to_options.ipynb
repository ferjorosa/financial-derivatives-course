# 1 - Options - A Financial Derivative

**What is an Option?**
- An option is a derivative financial instrument that specifies a contract between two parties.
- This contract pertains to a future transaction involving an asset at a reference price.
- The buyer of the option gains the right (but not the obligation) to engage in the transaction.
- The seller of the option incurs the obligation to fulfill the transaction.
- The key difference from futures is that one party in an options contract can choose whether the transaction takes place or not, depending on the price movement in their favor.

**Types of Options**
- Two main types of options: Call Options and Put Options.
- **Call Option**: The buyer has the right (not obligation) to buy a specified quantity of an underlying asset from the seller at a future date (expiration) for a certain price (strike price).
- **Put Option**: The buyer has the right (not obligation) to sell the asset at the strike price by the expiration date. The seller is obliged to buy the asset if the buyer exercises the option.
- Both buyers pay a premium for these rights.

**Buyer and Seller Perspectives**
- Call Option:
  - **Buyer**: Bullish, expects the underlying asset's price to rise.
  - **Seller**: Bearish, anticipates a stable or falling price.
- Put Option:
  - **Buyer**: Bearish, pays a premium, anticipates the price to fall.
  - **Seller**: Neutral or Bullish, hopes for a stable or rising price.

**American vs. European Options**
- **American Options**: Can be exercised at any time up to and including the expiration date.
- **European Options**: Can only be exercised on the expiration date.
- American options are more flexible but tend to be more expensive.

# 2 - Understanding Options: In the Money, At the Money, Out of the Money

**At the Money (ATM)**
- An option where the strike price matches the current price of the underlying asset.
- For example, if a stock is trading at $100, an option with a $100 strike price is referred to as "at the money."

**In the Money (ITM)**
- An option that is profitable for the buyer if exercised immediately.
- For a call option, the underlying's price is higher than the option's strike price (e.g., strike price is $100, and the stock is trading at $105).
- For a put option, the underlying's price is lower than the option's strike price (e.g., strike price is $100, and the stock is trading at $95).

**Out of the Money (OTM)**
- An option that has no intrinsic value and has not yet reached in or at the money status.
- For a call option, the underlying's price is lower than the option's strike price (e.g., strike price is $100, and the stock is trading at $95).
- For a put option, the underlying's price is higher than the option's strike price (e.g., strike price is $100, and the stock is trading at $105).
- OTM options need the underlying to move to become valuable at expiration.

**Intrinsic Value and Time Value**
- Intrinsic value: The amount by which an option is in the money.
  - For example, a call option with a $100 strike price and the stock at $105 has $5 of intrinsic value.
- Time value: The portion of an option's value beyond its intrinsic value.
  - Reflects the potential for further movement in the underlying.
  - An option priced at $6 with $5 intrinsic value has $1 of time value.
- The total option price is the sum of intrinsic value and time value.
- In the real world, options are worth more than just intrinsic value due to time value.


# 3 - Options Margin and Covered vs. Naked Options

**Introduction**
- Options trading involves various margin requirements, which are collateral funds traders must have in their accounts.
- Margin is crucial to ensure options positions are covered and settled.

**Options Margin**
- Options trading requires an initial deposit of funds in a margin account.
- **Margin is a security for the seller to fulfill their obligations**.
- Margin ensures that obligations are met and losses covered.
- Margin requirements vary depending on the broker and exchange rules.
- The amount of margin required is often a fraction of the option's price.

**Covered Options**
- **Covered Call**: In a covered call, the trader owns the underlying asset (e.g., stock) and sells a call option on that asset.
- Margin is generally not required for covered calls because the trader already holds the underlying asset.
- The existing asset "covers" the potential obligation.

**Naked Options**
- **Naked Call**: In a naked call, the trader sells a call option without owning the underlying asset.
- Margin is required for naked calls to ensure the trader can cover potential losses if the option is exercised.
- A naked call's risk is theoretically unlimited, making margin crucial for risk management.
- **Naked Put**: Similarly, selling a put option without holding the underlying asset is a naked put.
- Margin is also required for naked puts to cover potential obligations if the option is exercised.

**Key Differences**
- Covered options do not require margin because the trader already owns the underlying asset.
- Naked options, both calls and puts, require margin to cover potential losses.
- Naked options pose higher risk, while covered options offer some downside protection.


# 4 - Options Positions Overview

**Call Options**
- **Long Call Option (i.e., buying Call options)**: The buyer has the right (but not the obligation) to buy the underlying asset at an agreed-upon price at the expiration date.
  - Typically entered when the buyer expects the underlying asset's price to rise.
  - Premium is paid to acquire this right.
  - If the price rises, the buyer can exercise the right; if it doesn't, the buyer loses the premium paid.
<br></br>
- **Short Call Option (i.e., selling Call options)**: The seller has the obligation to sell the underlying asset.
  - Typically entered when the seller anticipates a fall in the value of the underlying or does not expect it to rise significantly.
  - The seller is bearish compared to the buyer.
  - Goal is to keep the premium received.

**Put Options**
- **Long Put Option (i.e., buying Put options)**: The buyer has the right (but not the obligation) to sell the underlying asset.
  - Typically entered when the buyer is bearish on the underlying asset and expects its price to fall.
  - The buyer pays a premium for this right.
  - Profits when the underlying asset's price falls below the premium paid.
<br></br>
- **Short Put Option (i.e., selling Put options)**: The seller has the opposite view to the buyer of a put option.
  - The seller anticipates the underlying asset's price to stay the same or rise.
  - The goal is to keep the premium received.

**Conclusion**

- A bit bearish: selling Call Options (usually covered)
- More bearish: buying Put Options
<br></br>
- A bit bullish: selling Put Options (usually covered)
- More bullish: buying Call Options


# 5 - Exercising Options

**What is Exercising an Option?**
- The owner or buyer of an option contract has the right, but not the obligation, to exercise the option.
- Exercising means carrying out the transaction specified in the contract immediately between the two parties.
- Exercising an option terminates the options contract.

**Exercising Call and Put Options**
- Call Option: The owner buys the underlying asset at the strike price from the option seller.
- Put Option: The owner sells the underlying asset to the option seller at the strike price.
- This is done only if the option is "in the money," as explained in a previous video.

**Option Style**
- The option style specified in the contract determines when, how, and under what circumstances the holder can exercise it.
- European-style options are exercisable only at the options expiration date.
- American-style options can be exercised at any time until expiration.
- Bermudan-style options are exercisable on specific dates.

**Manner of Settlement**
- Physically settled options require the actual delivery of the underlying asset (e.g., 100 shares of the underlying stock).
- Cash-settled options do not involve physical delivery; the cash difference is paid based on market value vs. strike price.
- Many exchange-traded index options are cash-settled due to the complexity of delivering a basket of stocks.

**OCC (Options Clearing Corporation)**
- In the United States, the OCC automatically exercises options set to expire in the money by one cent or more.
- This is done for broker convenience and is known as "exercised by exception."
- Holders can request not to exercise by exception.
- The exercise price is based on the last trade reported before 4:01 p.m. and 30 seconds on the day before expiration.
- The OCC tentatively reports the price at 4:15 p.m. and makes it official at 5:30 p.m.

# 7 - Understanding Put-Call Parity

**The Put-Call Parity Formula**
- The formula: Put + $S_{0}$ = Call + Present Value of the Strike ($K * e^{-rT}$).
  - $S_{0}$: underlying asset
  - $K$: strike price
- Combining options with the underlying creates the payoff of a different option.
- Must follow this formula; it's not just theoretical, but rather a market necessity to avoid arbitrage opportunities.

**Applying the Formula**
- The formula allows for various algebraic manipulations to derive related formulas.
- Traders use these formulas to determine option prices and identify arbitrage opportunities.

$S_{0} = c - p + K * e^{-rT}$

$p = c + K * e^{-rT} - S_{0}$

$c = S_{0} + p - K * e^{-rT}$

$K * e^{-rT} = S_{0} - c - p$

**Practical Example**
- Example: Underlying trading at $31, with put and call options both at a strike of $30.
- Interest rate: 10%, three-month call option at $3, put option at $2.25.
- Check for arbitrage opportunity by applying the formula.
- Portfolio A: Call + Present Value of the Strike = $3 + (30 * e^{-0.10 * 3/12}) = 32.26$.
- Portfolio B: Put + S0 = $2.25 + $31 = $33.25.
- Since Portfolio B is greater than Portfolio A, an arbitrage opportunity exists.
- To profit from it, buy the cheap portfolio (Portfolio A) and sell the expensive one (Portfolio B) for an arbitrage profit of $0.99.

**Conclusion**
- Put-call parity is a fundamental concept in options trading.
- It helps ensure that options with the same payoff trade at appropriate prices.


# 8 - Factors Affecting Option Prices

**Factors Affecting Option Prices**
1. **Current Stock Price:**
   - For call options, an increase in the underlying's price raises the call option's value.
   - For put options, a decrease in the underlying's price raises the put option's value.
<br></br>
2. **Strike Price:**
   - The difference between the strike price and the stock price at expiration affects an option's payoff.
<br></br>
3. **Time to Maturity:**
   - The more time remaining on an option, the more valuable it is.
   - Options with longer expiration periods are more valuable as more can happen during that time.
<br></br>
4. **Volatility:**
   - More volatile underlying assets lead to more valuable options.
   - Options on volatile assets have higher potential for large price swings.
<br></br>
5. **Risk-Free Interest Rate:**
   - Changes in risk-free interest rates affect option prices.
   - They impact present value and future cash flows in option pricing formulas.
<br></br>
6. **Dividends:**
   - Dividends paid by the underlying company affect option prices.
   - Dividends typically cause the stock price to drop by the dividend amount.

**Conclusion**
- Volatility is a critical factor in option pricing.
- Risk-free interest rates have an impact, but it's usually smaller.
- Dividends affect option prices due to their impact on underlying stock returns.


# 9 - Bull Call Spread

**What is a Bull Call Spread**
- A bull spread is designed to profit from a rise in the underlying security's price.
- Components of a bull call spread:
  - Buying a call option (strike price: K1).
  - Selling a call option (strike price: K2).
- Payoff diagram for a bull call spread:
  - It resembles a long call option's payoff until it reaches K2.
  - Once it hits K2, incremental gains are lost.

**Key Points**
- Bull call spreads cost less in premium compared to buying just a call option with strike K1.
- They reach the break-even point quicker.
- Maximum profit is the difference between the strike prices (K2 - K1) minus the premium paid.
- Maximum loss is the premium paid for the spread.

**Bear Call Spread (Opposite of Bull Call Spread)**
- A bear spread is designed to profit from a fall in the underlying security's price.
- It involves selling a call option (strike price: K1) and buying a call option (strike price: K2).
- The payoff of a bear call spread is the opposite of the bull call spread.

# 10 - Bear Put Spread

**What is a Bear Spread?**

- A bear spread is an option spread position.
- It involves **buying and selling equal amounts of the same class of options**.
- In this case, we're dealing with put options.
- The bear spread aims to profit from a fall in the price of the underlying asset.

**Components of a Bear Spread**

- The bear spread uses put options with the same expiration date but different strike prices.
- For example, buy an at-the-money put option and sell an out-of-the-money put option.
- The difference in strike prices determines the spread's potential profit.

**Payoff Profile**

- When the underlying asset's price falls, the bear spread begins to make money.
- The payoff profile resembles that of being long a put option until it reaches the lower strike price (k1).
- Beyond k1, the payoff starts to flatten.

**Why Use a Bear Spread?**

Reduced cost, less upside (although that upside is very unprobable, hence why is usually better to use a spread)

- Bear spreads are used to save money.
- It costs less to create a bear spread than to buy an individual put option with a higher strike price (k2).
- The trade-off is a reduced upside potential compared to owning the higher strike put option alone.
- Bear spreads can reach their break-even point earlier than individual put options.

**Maximum Gain and Loss**

- The maximum profit is the difference between the two strike prices (k2 - k1) minus the premium paid.
- The maximum loss is the premium paid for the spread (which is lower than the premium of a Put-only strategy).

**Implications**

- A bear spread profits from a moderate fall in the underlying asset's price.
- It's a cost-effective strategy compared to buying a single put option.
- Counterparties on the other side of the transaction have the opposite payoff.

**Put-Call Parity**

- Bear spreads can also be created with call options, thanks to put-call parity. For example, you can replicate the payoffs of a long call option by holding a long stock position, a short put option, and a risk-free bond.

# 11 - Butterfly Spreads


**Butterfly Spread Basics**
- Butterfly spread involves positions in three different options.
- Constructing a butterfly spread:
  - Buy two call options: one with a strike price of k1 and one with a strike price of k3.
  - Sell two call options with a strike price of k2 (in the middle).
- The combination creates a specific payoff structure.

**Strategy and Payoff**
- Butterfly spread is a limited risk, limited return strategy.
- Maximum profit occurs when the underlying asset expires at the strike price k2.
- It requires the underlying to close near k2 for profitability.
- Most other outcomes result in losses.

**Trading Directions**
- You can either buy or sell a butterfly spread.
- Buying the spread involves buying calls at k1 and k3 and selling two calls at k2.
- Selling the spread means selling calls at k1 and k3 and buying two calls at k2.
- Selling the spread results in the opposite payoff structure.

# 13 - Strangles

**Strangles vs. Straddles**
- Strangles are similar to straddles.
- In a straddle, you buy both a put and a call with the same strike price.
- In a strangle, you buy a put and a call with the same expiration date but different strike prices.
- Strangle components: an out-of-the-money put and an out-of-the-money call.
- Out-of-the-money options are cheaper than at-the-money options.

Basically both a straddle and a strangle bet on volatility, but the strangle is cheaper because you have less price "margin", you are betting further away

**Long vs. Short Strangle**
- You can buy or sell strangles.
- Buying a strangle means betting on high volatility.
- You buy both the put and the call.
- Not betting on the direction but on a significant move in the market.
- Selling a strangle means you think the underlying won't experience significant volatility.
- You sell both the put and the call, receiving premium.
- Breakeven points are on the upside and downside.
