# 1 - What are Financial Derivatives

**Introduction to Financial Derivatives**
- Financial derivatives are contracts derived from another asset or instrument.
- Used for risk management and speculation.

**Types of Derivatives**
- Includes futures contracts, options contracts, swaps, and forwards.
- Categorized by the relationship with the underlying asset, type of asset, and market.

**Purpose of Derivatives**
- Two primary purposes: speculation and hedging.
- Speculators aim to profit from price movements.
- Hedgers use derivatives to reduce existing risks.

**Types of Underlying Assets**
- Common assets include equities (stocks), fixed income (bonds), commodities, indices, forex, and non-traditional assets like weather events.

**Market Participants**
- Participants: hedgers, speculators, arbitrageurs, and middlemen.
- Hedgers are producers or consumers managing exposure.
- Speculators seek profit.
- Arbitrageurs exploit price differences.
- Middlemen facilitate trades for a commission.

**Arbitrage**
- Profiting from price differences in identical or similar instruments in different markets.
- Maintains prices close to fair value.

**Derivative Market Dynamics**
- Middlemen profit from high market activity through commissions or spreads.
- They hedge residual exposures to avoid significant profits or losses.

# 2 - How are Financial Derivatives Traded?

**How Derivatives are Traded**
- Two groups of derivatives: Over-the-Counter (OTC) and Exchange-Traded Derivatives.
- OTC derivatives are traded directly between two parties, subject to counterparty risk.
- OTC market primarily involves banks, large corporations, and sophisticated parties.
- Push to bring OTC derivatives onto clearing platforms for financial stability.
- Exchange-traded derivatives are traded via specialized derivatives exchanges.
- Exchanges act as intermediaries, taking margin payments for guarantee.
- Most exchange-traded derivatives are traded electronically through brokers or investment banks.

**Major Derivative Contract Types**
- The most common derivative contracts are:
  - **Forwards**: OTC contracts for future payment at a predetermined price.
  - **Futures**: Exchange-listed contracts for buying or selling assets on a future date at a specified price.
  - **Options**: Contracts that grant the buyer the right (but not the obligation) to transact at a predetermined price.
  - **Warrants**: Similar to call options, often issued by companies on their own stock.

**Clarification on Derivatives as Contracts**
- All derivatives are considered legal contracts between two parties.
- They do not represent ownership of underlying assets but rather contracts with payoffs based on the underlying's price.
- Derivatives markets are often misunderstood as legalized gambling, but they serve essential economic functions.
- They allow for risk allocation, stimulate the economy, and improve market efficiency.
- Derivatives can lead to large losses, primarily due to excessive leverage.

# 3 - What are Financial Futures

**What is a Future?**
- A futures contract is a **standardized** financial contract.
- Buyer is obligated to purchase, and the seller is obligated to sell an asset at a future date and price.
- Contracts specify details of the underlying asset, quantity, quality, and delivery.
- Settlement can be in cash or physical delivery.
- Price at which the transaction occurs is the futures price or strike price.
- Delivery and payment occur on a specified future date (delivery date).
- Negotiated on futures exchanges, acting as intermediaries.
- Parties: buyer (long) and seller (short).

**Historical Background**
- Futures contracts have a long history.
- First futures exchange was the Dojima rice exchange in Japan in the 1730s.
- Chicago Board of Trade listed the first standardized exchange-traded forward contracts in 1864.
- Chicago became a center for the derivatives market in the United States.

**Characteristics of Futures Contracts**
- Futures contracts are highly standardized.
- They specify:
  - Underlying asset or instrument (commodities, currencies, securities).
  - Settlement type (cash or physical).
  - Quantity and units per contract.
  - Quoted currency.
  - Deliverable grade (specific quality criteria).
  - Last trading time and date.
  - Delivery date and location.
  - Tick size (minimum price fluctuation).

# 4 - What are Forward Contracts?

**What is a Forward?**
- Forward contracts are financial agreements to buy or sell an asset at a future date for a predetermined price.
- Similar to futures contracts but **not standardized**; customized to the parties involved.
- Common in various markets, including commodities, currencies, and interest rates.

**Characteristics of Forward Contracts**
- Customized: Parties can tailor terms to their specific needs.
- No standardization: Lack of uniformity makes them unique to each transaction.
- Private agreements: Typically traded over-the-counter (OTC) rather than on exchanges.
- Flexibility: Parties can agree on various aspects, such as quantity, price, and delivery date.
- Settlement: Often settled in cash but can involve physical delivery.

**Key Terms in Forward Contracts**
- **Underlying Asset**: The asset being bought or sold in the contract (e.g., currency, commodity).
- **Delivery Date**: The future date when the contract is settled.
- **Delivery Price**: The price at which the asset will be bought or sold.
- **Long Position**: The party agreeing to buy the asset.
- **Short Position**: The party agreeing to sell the asset.
- **Mark-to-Market** (Optional): Tracking the contract's value over time.

**Advantages and Disadvantages**
- **Advantages**:
  - Customization: Tailored to specific needs.
  - Flexibility: Parties have control over contract terms.
  - Privacy: OTC trading keeps details confidential.
- **Disadvantages**:
  - Counterparty Risk: Default risk if one party fails to meet obligations.
  - Lack of Liquidity: Not as easily tradable as standardized futures.
  - Complexity: Customization can lead to complex contracts.

**Conclusion**
- Forward contracts provide flexibility for parties to manage risk or speculate.
- Customization and private nature distinguish them from standardized futures.
- Used across various markets and industries.
- Important tool for managing price risk and achieving financial goals.

# 5 - Payoff Diagrams for Futures and Forwards

Understanding Payoff Diagrams for Futures and Forwards
- Futures and forwards have straightforward payoff diagrams.
- Payoff diagrams help beginners understand derivatives.
- These simple diagrams make understanding more complex ones (e.g., options) easier.
- First, let's examine payoff diagrams for being long or short the underlying asset.

**Long vs. Short**
- In finance, being "long" means owning an asset.
- Being "short" involves borrowing the asset and selling it, intending to buy it back later.
- **Long position**: Profit when the asset price rises, losses when it falls, linear relationship. 
  - Profits are unlimited as asset prices can rise indefinitely.
  - Losses are limited, as prices cannot go below zero.
- **Short position:** Profit when the asset price falls, losses when it rises, linear relationship.
  - Gains are capped when asset prices fall to zero.
  - Losses are unlimited as prices can keep rising.

**Futures and Forwards Payoff Diagrams**
- Payoff profiles for futures and forwards are nearly identical to those for being long or short the underlying asset.
- Investors enter at the futures price and the ultimate payoff depends on the final futures price (expiration or contract closure).
- Benefits of trading derivatives like futures: For every buyer, there's a seller, so no need to borrow the underlying asset.
- Simply sell a futures contract to someone with the opposite position.

# 6 - What is Futures Margin and How does it work?

**Understanding Margin**
- Margin refers to the money used to buy securities, often referred to as "buying on margin."
- It's the amount of equity contributed by a customer as a percentage of the current market value of securities.
- A 50% margin requirement means you need to put up half the money to buy the securities.
- Exchanges and counterparties demand margin to minimize credit risk, ensuring traders can meet their obligations.
- Traders must post margin, often a liquid bond or performance bond, typically 5% to 15% of the contract's volume.
- Futures margin rates are set by exchanges, with brokerages sometimes adding a premium.
- Margin rates are usually based on risk; higher-risk contracts have higher margins.
- To minimize counterparty risk, trades on regulated futures exchanges are guaranteed by clearinghouses.

**Types of Margin**
- **Initial Futures Margin**: Money required to initiate a futures contract position.
  - Calculated based on the maximum estimated change in contract value within a trading day.
  - Set by the exchange.
- **Margin Maintenance**: Minimum margin per outstanding futures contract to be maintained in the margin account.
  - Maintenance margin is usually lower than initial margin, providing some flexibility.
- **Margin Call**: Triggered when the account value drops below the maintenance level.
  - Broker asks the client to deposit sufficient funds to maintain their trading position, including maintenance margin.
  - Margin calls require immediate action.

# 7 - Futures Clearinghouse

**Understanding Futures Clearinghouse**
- Futures exchanges are regulated and trades are guaranteed by a Clearing House.
- The Clearing House is separate from traders and brokerages/market makers.
- Purpose: Minimize counterparty risk, ensure trade settlement.
- Clearing Houses become the counterparty for all trades they clear.
- Reduce risk of clearing firms failing to meet trade settlement obligations.
- Responsible for settling accounts, clearing trades, collecting margin, regulating delivery, and reporting trading data.
- Acts as a third party to all futures and options contracts, ensuring trades happen as planned.
- Key risk reduction measures:
  1. Netting offsetting transactions between counterparties.
  2. Requiring collateral margin deposits.
  3. Providing independent valuation of trades and collateral.
  4. Monitoring the creditworthiness of clearing firms.
  5. Guarantee fund to cover losses exceeding defaulting firms' collateral.
- Clearing fees: Charges assessed by Clearing Houses for using their facilities.
- Transaction costs in futures trading include brokerage fees and clearing fees.
- Delivery fees may be included, but actual delivery in futures contracts is rare.
- Clearing fees vary based on transaction type and size.

**Futures Clearing Merchant (FCM)**
- FCM is a brokerage firm that is an exchange member.
- Acts as an intermediary between investors, futures exchanges, and the Clearing House.
- Ensures proper trade settlement and transaction success.
- Responsibilities include holding customer funds, clearing trades, and back-office functions.
  - Marking to market customer's futures account.
  - Sending trade confirmations, account summaries, and year-end tax forms.


# 8 - Pricing Financial Futures

**Pricing Futures - Intuition**
- Imagine there's no formula to determine fair future prices.
- Example: Share price currently at $40, futures available for the stock.
- Futures broker quotes a futures contract at $43, **expiring in three months**.
- Interest rate is 5%.
- We want to determine if $43 is a fair price for the futures contract.

**Scenario #1**
- Buy one share for $40.
- Borrow $40 at 5% for three months.
- Sell a futures contract on one share at $43.
- You are **long with respect to the stock** because you bought one share of the underlying stock. You hedged by selling the futures contract.
- Results: We earn $43 and owe $40.50.
- Net profit: $43 - $40.50 = $2.50 per contract.
- Understanding interest: We owe interest because money has value over time.


**Scenario #2**
- Buy futures contract at $39.
- Short sell the share for $40.
- You are **short with respect to the stock** because you shorted (sold) one share of the underlying stock. You hedged by buying the futures contract.
- Interest earns $0.50.
- Results: Earn $40.50 and owe $39.
- Net profit: $40.50 - $39 = $1.50 per contract.


**Conclusion**
- Fair value of the futures contract: $40.50.
- No need for a formula; we can deduce the fair value intuitively. If the value was other than $40.50, people would be trading the contracts for "free money".
- Real-world factors may influence futures prices, but the basic concept remains.
- [Use continuously compounded interest with Euler's number (2.718) for calculations](https://www.cuemath.com/continuous-compounding-formula/). The resulting Future's price formula is: $$F_{0} = S_{0} * e^{RT} = 40 * 2.718^{0.05*1/4}$$
    where $F_{0}$ is the price of the future's contract, $S_{0}$ is the price of the underlying asset, $R$ is the rate of interest, and $T$ is the time


**Real-World Considerations**
- In practice, real-world factors like transaction costs, borrowing/lending rates, and short-selling restrictions affect futures prices.
- These factors prevent complete arbitrage.
- Futures prices may vary within arbitrage boundaries around the theoretical price.

**Next**
- Next we will explore exceptions to the basic formula.
- It will cover assets with storage costs, assets with cash flows, assets that can't be borrowed, and shortages of assets during the contract's life.
- Modifications to the formula will address these exceptions.



# 9 - Pricing Financial Futures (Exceptions to the basic formula)

**Pricing Futures with Cash Flow**
- In a perfect market, futures pricing is based on a formula.
- Real-world factors like transaction costs and borrowing rates can create variations.
- Modifications needed for cases with cash flows associated with the underlying asset (e.g., interest rates on bonds, dividends).
- Formula: $$F_{0} = S_{0}  * e^{(R-Q) * T}$$
  - $F_{0}$: Fair value of future at time 0
  - $S_{0}$: Price of the underlying asset at time 0
  - $R$: Rate of iterest
  - $Q$: Percentage cash flow on the underlying asset.
  - $T$: Time remaining until expiration.

**Dealing with Storage Costs**
- Some underlyings, like physical commodities (e.g., oil, gold), have associated storage costs.
- Storage costs are negative cash flows.
- Formula for pricing futures with storage costs: $$F_{0} = S_{0}  * e^{(R + STC) * T}$$
  - STC: Percentage storage cost on the underlying.

**Pricing Foreign Currency Futures**
- Foreign currency futures involve two interest rates: local (domestic) and foreign currency interest rates.
- Formula for pricing foreign currency futures: $$F_0 = S_0 * e^{(R - RF) * T}$$
  - $R$: Local (domestic) interest rate.
  - $RF$: Foreign currency interest rate.


# 10 - Convenience Yield in Futures Market

**Understanding Convenience Yield**
- Convenience yield refers to the benefit derived from owning a consumption asset.
- Consumption assets are used in daily life, in manufacturing, or are essential to businesses.
- Users of these assets prefer owning the actual asset over owning a futures or forward contract.
- Classic example: Heating oil in winter - people would pay more for the physical oil than for a future contract that delivers later when not needed.

**Convenience Yield Explanation**
- Convenience yield represents the benefit of owning the actual asset instead of its derivative.
- Owners of inventory choose between consumption today or investment for the future.
- Rational individuals choose the option that best suits their needs.
- Example: The cost of a Christmas tree closer to Christmas vs. after Christmas.
- Example: The cost of owning physical oil in winter versus owning a contract for buying oil in the future.

**Calculating Convenience Yield**
- Convenience yield is backed out from market prices.
- You can't have a fixed number for convenience yield because it varies.
- <span style="color:blue"><b>Convenience yield is not a fixed number; it's market-implied at a specific time</b></span>

**Adjusting Futures Pricing Formula**
- Futures pricing formula: $$F_{0} = S_{0}  * e^{(R + STC - Y) * T}$$
- $R$: Interest rate.
- $STC$: Storage costs (if applicable).
- $Y$: Convenience yield.
- $T$: Time to expiration of the futures or forward contract.


## 11 - Understanding Backwardation and Contango in Futures Markets

**Introduction**
- Backwardation and contango are terms used in futures markets.
- Background: John Maynard Keynes believed commodity producers are natural hedgers.
- Hedgers want to sell commodities in the future to reduce risk.
- If hedgers collectively hold a net short position in futures markets, speculators hold the other side.
- Speculators must hold a net long position.

**Backwardation:**
- Backwardation is a market situation where <span style="color:blue">the futures price of a commodity is less than the expected future spot price of that commodity at the contract's maturity</span>.
- This condition suggests that the futures market is pricing the commodity at a discount compared to the expected future market price.
- Backwardation typically occurs when there is a strong demand for the commodity in the present, and this demand drives up its spot price, making it higher than the futures price.
- It is often associated with commodities that are in short supply or facing immediate delivery constraints.
- For hedgers, backwardation can be favorable as it allows them to lock in a sale price for their commodity that is higher than the current futures price.

**Contango**
- Contango, on the other hand, is a market condition where <span style="color:blue">the futures price of a commodity is higher than the expected future spot price at the contract's maturity</span>.
- In contango, the futures market prices the commodity at a premium compared to the expected future market price.
- Contango is often driven by factors like storage costs, financing costs, or expectations of oversupply in the future.
- For commodities with carrying costs (like storage, insurance, and financing), it is common for futures prices to be higher than spot prices.
- Speculators who are long in a contango market may face the challenge of holding the commodity at higher costs.

**Factors Influencing Backwardation and Contango**
- Various theories exist for why certain products trade in backwardation or contango.
- It's still a puzzle that academics try to explain.
- Some commodities usually trade in one condition but may flip to the other due to supply and demand pressures.
- These market conditions can significantly impact a trader's profit and loss (P&L).


# 12 - Hedging Strategies for Companies

**Foreign Exchange Risk Example**
- Scenario: An American manufacturing company.
- They source supplies locally in U.S. dollars, make products in the U.S., sell to U.S. consumers, and pay American staff in U.S. dollars.
- The question is, does this company have foreign exchange risk?
- Answer: Yes, if there's a foreign competitor.
- If the foreign currency falls relative to the U.S. dollar, the foreign competitor gains a cost advantage and can undercut prices, affecting the U.S. company.

**Assessing the Need to Hedge**
- Every company is exposed to certain risks, including foreign exchange risk.
- Should a company hedge this risk?
- It may not be productive or necessary in some cases, especially for simpler goods manufacturing.
- Being aware of risks is crucial.

**Why Companies Hedge**
- Companies may lack expertise in predicting market prices, such as interest rates, exchange rates, and commodity prices.
- If these prices significantly impact costs, hedging can reduce business uncertainty.
- It allows companies to focus on their core expertise.

**Counterargument: Individual Shareholder Hedging**
- Some argue that shareholders can hedge market exposures individually.
- Companies shouldn't hedge because investors may have specific reasons for investing (e.g., in a gold mining company) and may not want their gains affected by hedging.
- It is more cost-effective to do it at a company level due to "size discounts".

**Hedging for Visibility and Planning**
- Companies that hedge gain visibility into future prices.
- It aids in making long-term projections and business plans with stable profit margins.
- Example: A breakfast cereal company decides the price of a box of oatmeal a year in advance. Hedging helps them have stable pricing for supermarkets and customers.

**Types of Futures Hedges**
- Short futures hedges are used when a company owns the asset and plans to sell it in the future (e.g., farmers, gold mining companies). 
  - **Protects against future falling prices** of the underlying (usually a comodity).
- Long futures hedges are used when a company knows it will have to purchase the asset in the future (e.g., a breakfast cereal company buying grains, or an airline needing jet fuel). 
  - **Protects against future rising prices** of the underlying (usually a comodity).



# 13 - Types of Orders in Futures Trading

**Market Order**
- The simplest order type.
- Instructs a broker to buy or sell at the prevailing market price.
- Guarantees execution but may not guarantee the desired price.
- Useful for fast execution but carries price uncertainty.

**Limit Order**
- Popular among traders.
- Specifies the price at which a trader is willing to buy (no more than) or sell (no less than).
- Offers price control but doesn't guarantee execution.
- Traders use limit orders to manage the price they're willing to accept.

**Stop Order (Stop-Loss Order)**
- Triggers a market order to buy or sell when the price reaches a specified level (the stop price).
- Buy stop orders are placed above the current market price, sell stop orders below it.
- Used to limit losses or protect profits.
- Some exchanges have stopped offering stop orders due to perceived risks.

**Stop-Limit Order**
- Combines features of a stop order and a limit order.
- Functions as a stop order initially but, when triggered, becomes a limit order.
- Specifies the limit price to buy or sell at no more or less than the limit.

**Market If Touched Order**
- Becomes a market order if the market price touches a specified level.
- Executed at the best available price once the trigger is touched.

**Discretionary (Market-Not-Held) Order**
- Relies on the broker's discretion to obtain the best available price.
- The broker attempts to secure the most favorable price.

#### 1 - Show that the minimum variance hedge ratio ($h$) when trying cross-hedge a portfolio's volatility using overall market Futures is equal to $\beta$

Just to remember, the minimum variance hedge ratio ($h$) has the following form: 
$$h =  \frac{\rho_{E,F} \ \sigma_{E}}{\sigma_{F}}$$ 

where 
* $E$ refers to the asset we are exposed to (e.g., a variety of oil, a stock portfolio, etc.), 
* $F$ refers to the asset we want to use its futures to hedge with (e.g., a different variety of oil whose futures are more liquid, the futures of the overall market, etc.), 
* $\rho_{E,F}$ is the Pearson correlation between $E$ and $F$
* $\sigma$ is the standard deviation

Now, the formula for Pearson's correlation is as follows:

$$ \rho_{E,F} = \frac{\text{cov(E, F)}}{\sigma_{E}\sigma_{F}}$$

If we expand $h$ using this information we end up with:

$$ h = \frac{\text{cov(E, F)}\sigma_{E}}{\sigma_{E}\sigma_{F}\sigma_{F}} = \frac{\text{cov(E, F)}}{\sigma_{F}^{2}}$$

If we consider $E$ to be our portfolio and $F$ to be the overall market, we have that $h = \beta$. In that case, we can hedge our full portolio by buying N overall market futures contracts:

$$ N = \frac{\beta \ V_{E}}{V_{F}} $$

where 

* $V_{E}$ refers to the total value of our exposure (e.g., our portfolio)
* $V_{F}$ refers to the value of one futures contract of our cross-hedge asset (e.g., a futures contract of the overall market)


#### 2 - You manage a stock portfolio worth 100 million dollars. The portfolio has a $\beta$ of 1.1. Explain how would you fully hedge this portfolio if the SP500 futures contract is currently trading at 2000

Well, first we need to remember that the formula of $\beta$ is as follows: $\frac{\text{cov(E, F)}}{\sigma_{F}^{2}}$ and that the number of futures contracts that we need to fully hedge our portfolio comes from the following formula: $ N = \frac{\beta \ V_{E}}{V_{F}} $.

In this case, we know that:
* $\beta = 1.1$
* V_{E} = 100,000,000
* V_{F} = 2000

where

* $V_{E}$ refers to the total value of our exposure (e.g., our portfolio)
* $V_{F}$ refers to the value of one futures contract of our cross-hedge asset (e.g., a futures contract of the overall market)

Therefore, we would need 55000 contracts (assuming our beta refers to the SP500):

$$ N = \frac{\beta \ V_{E}}{V_{F}} = \frac{1.1 * 100,000,000}{2000} = 55000$$
